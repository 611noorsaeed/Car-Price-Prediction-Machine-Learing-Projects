# Extracting Training Data

```python
X = car[['name','company','fuel_type','year']]
y = car['Price']
```

# Applying Train Test Split

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import make_column_transformer
from sklearn.pipeline import make_pipeline
```

# Creating an OneHotEncoder object to contain all the possible categories

```python
ohe = OneHotEncoder()
```

```python
ohe.fit(X[['name','company','fuel_type']])
```

```python
col_trans = make_column_transformer((OneHotEncoder(categories=ohe.categories_),['name','company','fuel_type']),remainder='passthrough')
```

# Linear Regression Model with pipe

```python
lr = LinearRegression()
```

```python
pipe = make_pipeline(col_trans,lr)
```

```python
pipe.fit(X_train,y_train)
```

```python
ypred = pipe.predict(X_test)
```

```python
from sklearn.metrics import r2_score
```

```python
r2_score(y_test,ypred)
```

# Finding the model with a random state of TrainTestSplit where the model was found to give almost 0.92 as r2_score

```python
scores = []
for i in range(1000):
    X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.1,random_state=i)
    lr = LinearRegression()
    pipe =make_pipeline(col_trans,lr)
    pipe.fit(X_train,y_train)
    ypred=pipe.predict(X_test)
    scores.append(r2_score(y_test,ypred))
scores
```

```python
# finally best model 
import numpy as np

X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.1,random_state=np.argmax(scores))
lr=LinearRegression()
pipe=make_pipeline(col_trans,lr)
pipe.fit(X_train,y_train)
y_pred=pipe.predict(X_test)
r2_score(y_test,y_pred)
```
